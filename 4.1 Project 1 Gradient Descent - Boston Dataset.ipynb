```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("4.1 0000000000002417_training_boston_x_y_train .csv")
```

```python
data.shape
```

```python
def step_gd(X,Y,lr,theta):
    M = len(X)
    temp = Y - np.dot(X , theta)
    temp = np.dot(X.T , temp)
    temp = (-2/M)*temp
    theta = (theta - lr*temp)
    return theta
```

```python
def cost(X,Y,theta):
    M = len(X)
    return (1/M)*((Y - np.dot(X,theta))**2).sum()
```

```python
def gd(X,Y,lr,n_itr):
    theta = np.zeros(X.shape[1]).reshape(X.shape[1],1)
    for i in range(n_itr):
        theta = step_gd(X,Y,lr,theta)
        print(cost(X,Y,theta))
    return theta
```

```python
def run():
    Y= data.iloc[:,13]
    X= data.iloc[:,0:13]
    y = Y.values.reshape(X.shape[0],1)
    X['1'] = X[' ZN'] - X[' ZN'] + 1
    learning_rate = 0.1
    n_iterations = 100
    theta = gd(X,y,learning_rate,n_iterations)    
    return theta    
```

```python
theta = run()
```
